## Perguntas 
1. Crescimento entre seguradoras por premio ganho e emitido
2. top premio emitido
3. Market Share regra do link
4. Indice de Eficiencia : Despesa Administrativa/premio ganho
5. Indice de Eficiencia : Despesa Administrativa/premio emitido
6. indice de sinistralidade: sinistros ocorridos / premio ganho
7. premio ganho + premio ganho risco / custo de aquisicao risco + custo de aquisicao normal
8. combined: sinistralidade + custo de aquisicao / premio ganho

```python
#### COMPARATIVO CONSIDERANDO OS RAMOS DE VIDA E VGBL ####

#pega as bases utilizadas
balanco = SES_balanco.copy()
comparativo_ramo = SES_premios.copy()

balanco = balanco.loc[(balanco['damesano']>201612) * (balanco['quadro'] == '23 ')] #quadro 23 são os dados da DRE
balanco = pd.pivot_table(comparativo, index=['Seguradora','Ano'],columns='codCampo', values='valor').reset_index()
comparativo_ramo = comparativo_ramo.loc[comparativo_ramo['damesano']>201612]


#dados de despesa adminstrativa
da_balanco = balanco[balanco.cmpid==4069]
da_balanco ['damesano'] = pd.to_datetime(da_balanco['damesano'],format='%Y%m')
da_balanco17_19 = da_balanco.loc[da_balanco['damesano'].dt.month == 12]
da_balanco20 = da_balanco.loc[da_balanco['damesano'] == '2020-05' ]
da_balanco =da_balanco17_19.append(da_balanco20, ignore_index=True)
da_balanco.damesano = da_balanco.damesano.dt.year
da_balanco.rename(columns={"valor": "da"},inplace=True )
da_balanco.drop(['cmpid','quadro','seq'],axis=1,inplace=True)

#dados de vgbl
vgbl_balanco = balanco[balanco.cmpid.isin([7186,6238,6256])]
vgbl_balanco ['damesano'] = pd.to_datetime(vgbl_balanco['damesano'],format='%Y%m')
vgbl_balanco17_19 = vgbl_balanco.loc[vgbl_balanco['damesano'].dt.month == 12]
vgbl_balanco20 = vgbl_balanco.loc[vgbl_balanco['damesano'] == '2020-05' ]
vgbl_balanco =vgbl_balanco17_19.append(vgbl_balanco20, ignore_index=True)
vgbl_balanco.damesano = vgbl_balanco.damesano.dt.year
vgbl_balanco = vgbl_balanco.groupby(['coenti','damesano']).sum()[['valor']].reset_index()
vgbl_balanco.rename(columns={"valor": "premio_vgbl"},inplace=True )

#dados de risco previ
risco_balanco = balanco[balanco.cmpid==7220]
risco_balanco ['damesano'] = pd.to_datetime(risco_balanco['damesano'],format='%Y%m')
risco_balanco17_19 = risco_balanco.loc[risco_balanco['damesano'].dt.month == 12]
risco_balanco20 = risco_balanco.loc[risco_balanco['damesano'] == '2020-05' ]
risco_balanco =risco_balanco17_19.append(risco_balanco20, ignore_index=True)
risco_balanco.damesano = risco_balanco.damesano.dt.year
risco_balanco.rename(columns={"valor": "premio_risco"},inplace=True )
risco_balanco.drop(['cmpid','quadro','seq'],axis=1,inplace=True)

#dados de premios por ramo
comparativo_ramo ['damesano'] = pd.to_datetime(comparativo_ramo['damesano'],format='%Y%m')
comparativo_ramo.damesano = comparativo_ramo.damesano.dt.year 
todos_ramo = comparativo_ramo
comparativo_ramo = comparativo_ramo[(comparativo_ramo.coramo.astype(int).astype(str).str[:-2].astype(int).isin([9,13])) | (comparativo_ramo.coramo==1198) ]
comparativo_ramo = comparativo_ramo.merge(SES_cias,left_on='coenti',right_on='Coenti',how='inner')

#dados de grupo
chave_grupos_empresa = Ses_grupos_economicos.groupby('coenti').max().reset_index()[['coenti','damesano']]
grupos_empresa = Ses_grupos_economicos.merge(chave_grupos_empresa,how='inner',left_on=['coenti','damesano'],right_on=['coenti','damesano'])
grupos_empresa.nogrupo = np.where(grupos_empresa.cogrupo.isin(['1225','99999']), grupos_empresa.noenti.str.strip(),grupos_empresa.nogrupo)
grupos_empresa = grupos_empresa[['coenti','nogrupo']]
grupos_empresa.replace(['Zurich Santander Brasil Seguros e Previdencia S/A','ZURICH BRASIL COMPANHIA DE SEGUROS'],'ZURICH',inplace=True)
grupos_empresa.nogrupo = grupos_empresa.nogrupo.str.strip()


colunas_interesse = ['premio_direto','premio_ganho','sinistro_ocorrido','desp_com']#,'premio_emitido_cap']
for var in colunas_interesse:
    comparativo_ramo[var] = pd.to_numeric(comparativo_ramo[var].str.replace(',','.'))
    todos_ramo[var] = pd.to_numeric(todos_ramo[var].str.replace(',','.'))
comparativo_ramo = comparativo_ramo.groupby(['coenti','Noenti','damesano']).sum().reset_index()

#dados de premios para todos os ramos
todos_ramo = todos_ramo.groupby(['coenti','damesano']).sum()[['premio_ganho']].reset_index()
todos_ramo.rename(columns={"premio_ganho": "premio_ganho_todos_ramos"},inplace=True )


#juntando as bases
comparativo_ramo = comparativo_ramo.merge(da_balanco,left_on=['coenti','damesano'],right_on=['coenti','damesano'])
comparativo_ramo = comparativo_ramo.merge(vgbl_balanco,left_on=['coenti','damesano'],right_on=['coenti','damesano'])
comparativo_ramo = comparativo_ramo.merge(todos_ramo,left_on=['coenti','damesano'],right_on=['coenti','damesano'])
comparativo_ramo = comparativo_ramo.merge(grupos_empresa,left_on='coenti',right_on='coenti')
comparativo_ramo = comparativo_ramo.merge(risco_balanco,left_on=['coenti','damesano'],right_on=['coenti','damesano'])

#criando indicadores
comparativo_ramo.premio_ganho = comparativo_ramo.premio_ganho + comparativo_ramo.premio_risco
comparativo_ramo.premio_ganho_todos_ramos = comparativo_ramo.premio_risco + comparativo_ramo.premio_ganho_todos_ramos
proporcao_receita = (comparativo_ramo.premio_ganho + comparativo_ramo.premio_vgbl) / (comparativo_ramo.premio_ganho_todos_ramos + comparativo_ramo.premio_vgbl)
premio_ganho_total = comparativo_ramo.premio_ganho + comparativo_ramo.premio_vgbl
comparativo_ramo['da_proporcional'] = comparativo_ramo.da * proporcao_receita
comparativo_ramo['IEO'] = (-comparativo_ramo.da_proporcional / premio_ganho_total) * 100
comparativo_ramo['IEO_inteira'] = (-comparativo_ramo.da / premio_ganho_total) * 100
comparativo_ramo['CA/Pe'] = (comparativo_ramo.desp_com / comparativo_ramo.premio_direto) *  100
comparativo_ramo['combined'] = ((comparativo_ramo.desp_com + comparativo_ramo.sinistro_ocorrido) / comparativo_ramo.premio_ganho )*  100
comparativo_ramo['Sinistralidade'] = (comparativo_ramo.sinistro_ocorrido / comparativo_ramo.premio_ganho) * 100


comparativo_ramo = comparativo_ramo[comparativo_ramo.premio_ganho>0]
comparativo_ramo.drop(['cogrupo','coramo','Coenti','Nogrupo'],axis=1,inplace=True)
#'desp_com', 'sinistro_ocorrido', 'CA/Pe', 'combined', 'Sinistralidade', 'premio_direto'
comparativo_ramo = comparativo_ramo[['coenti','nogrupo', 'Noenti', 'damesano', 'premio_ganho','premio_risco','da', 'premio_vgbl', 'premio_ganho_todos_ramos', 'da_proporcional', 'IEO', 'IEO_inteira']]
comparativo_ramo.to_excel('G:/04 - Gerência de Produtos e Processos de Seguros/4.3 Informações Gerenciais/6. Mercado Segurador/SES-Susep/comparativo_ramo.xls',float_format='%.2f',inf_rep='',merge_cells=False,index=False)

```
